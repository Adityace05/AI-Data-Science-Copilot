```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(".//CSV//titanic.csv")
df.shape
```

```python
numerical_cols = df.select_dtypes(
    include=['int64','float64']
).columns.tolist()

categorical_cols = df.select_dtypes(
    include=['object']
).columns.tolist()

print("Numerical:", numerical_cols)
print("Categorical:", categorical_cols)
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df['Age'].dropna(),
    bins=20
)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

```python
for col in numerical_cols:

    plt.figure(figsize=(8,5))

    plt.hist(
        df[col].dropna(),
        bins=20
    )

    plt.title(f"{col} Distribution")

    plt.xlabel(col)

    plt.ylabel("Frequency")

    plt.show()
```

```python
for col in numerical_cols:

    plt.figure(figsize=(8,5))

    sns.boxplot(
        x=df[col]
    )

    plt.title(
        f"{col} Boxplot"
    )

    plt.show()
```

```python
for col in categorical_cols:

    if df[col].nunique() <= 10:

        plt.figure(figsize=(8,5))

        sns.countplot(
            x=df[col]
        )

        plt.title(
            f"{col} Countplot"
        )

        plt.xticks(rotation=45)

        plt.show()
```

```python
corr_matrix = (
    df.select_dtypes(
        include=['int64','float64']
    )
    .corr()
)
```

# Save Visualizations

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f"
)

plt.title(
    "Correlation Heatmap"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df['Age'].dropna(),
    bins=20
)

plt.title(
    "Age Distribution"
)

plt.savefig(
    "age_distribution.png"
)

plt.close()
```
